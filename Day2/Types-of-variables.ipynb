### Global Variables
- Variables that are created outside the function.
- We can use by everywhere, inside and outside function.
- If we create the same name of variable inside the function that variable is local which can be used inside the function only.The global variable with the same name will remain as it was, global and with the original value. 

````python
x = "fun" # Global variable

def myfunction(): # User create function
    print("Pyhtn is " + x)
    
myfunction() # function call

def fun():
  x = "Aweson" # local variable 
  print("Python is  " + x)

fun() # function call
print("Python is " + x)


````

### The global Keyword
- global keywors is used to creat a global variable inside a function.
- It is also used to change the value of global variable inside a function.

````python
def myfun():
    global x # x acts as global variable
    x = "Physco"
    print(x)

myfun()
print("Listining music " + x)
````
