### Numbers
    There are three numereic types in python

- int : int or interer is a whole number, positive or negative without decimal.
``` 
x = 1
y = 35656222554887711
z = -3255522
``` 
- float : Float or "floating point number" is a positive , negative number having one or more decimals. 
```
x = 1.10
y = 1.0
z = -35.59
```
    - It can be sintific number.

```
x = 35e3
y = 12E4
z = -87.7e100
```

- Complex : Complex numbers are written with "j" as  the imaginary part.
```
x = 3+5j
y = 5j
z = -5j
```

````python
x = 1
y = 35656222554887711
z = -3255522
print(type(x))
print(type(y))
print(type(z))



x = 1.10
y = 1.0
z = -35.59
print(type(x))
print(type(y))
print(type(z))

x = 35e3
y = 12E4
z = -87.7e100
print(type(x))
print(type(y))
print(type(z))

x = 3+5j
y = 5j
z = -5j
print(type(x))
print(type(y))
print(type(z))
````
