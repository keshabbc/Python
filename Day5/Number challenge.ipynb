### Numbers
- Create a variable x and assign it the integer 5
- Create a variable y and assign it the float 3.14
- Create a variable z and assign it the complex number 2+3j
- Print the type of each variable using type()

```python
x = 5
y = 3.14
z = 2+3j

print(type(x))
print(type(y))
print(type(z))
```
