### Data Types
     Variables can store different data. Data are of different types.
- Text Types:  str
- Numberic Types:  int,float,complex
- Sequence Type: list,range,tuple 
- Mapping Type: dict
- Set Type: set,frogenset
- Boolean Types: bool
- Binary Types:  bytes,bytearray,memoryview
- None Type: Nonotype   

        We can get the data type of any object by using the type() function.
        
    - Data type is set when we assign the value to a variable.

        ```
        x = 6
        print(type(x))
        ```



````python
x = 6
print(type(x))
````

````python
x = "Hello World"
print(type(x))
x = 20
print(type(x))
x = 20.5
print(type(x))
x = 1j
print(type(x))
x = ["Apple", "Banana", "Cherry"]
print(type(x))
x = ("apple", "Banana", "Cherry")
print(type(x))
x = {"Apple", "Banana", "Cherry"}
print(type(x))
x = frozenset({"Apple", "Banana", "Cherry"})
print(type(x))
x = True
print(type(x))
x = b"Hello"
print(type(x))
x = bytearray(5)
print(type(x))
x = memoryview(bytearray(5))
print(type(x))
x = None
print(type(x))
````

````python
# To specify the data types we use construction function.

x = int(10) 
print(x)
print(type(x))

x = str("Hello World")
print(x)
print(type(x))

x = float(20.7)
print(x)
print(type(x))

x = complex(1j)
print(x)
print(type(x))

x = list(("Apple", "Banana", "Cherry"))
print(x)
print(type(x))

x = tuple(("Apple", "Banana", "Cherry"))
print(x)
print(type(x))

x = frozenset(("Apple", "Banana", "Cherry"))
print(x)
print(type(x))

x = set(("Apple", "Banana", "Cherry"))
print(x)
print(type(x))

x = range(9)
print(x)
print(type(x))

x = bool(7)
print(x)
print(type(x))

x = bytes(9)
print(x)
print(type(x))

x = bytearray(9)
print(x)
print(type(x))

x = memoryview(bytes(9))
print(x)
print(type(x))

````
