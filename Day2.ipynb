# Variables
- Variable is a container used to store data values.
- In python there is no command to declare a variable.
- Variable is created when we assign a value.
- Name of variable is case sensative.
- To specify data types of a variable we use casting.
- String variable can be declare by using single or double quotes.
- To find the data type of a variable we use type() 



````python
# example of variable
x = 5          # here x is a variable
y = "Keshab"   # y is a string variable
print(x)
print(y)
````

### Casting:
- To specify the data types of a variable we use casting.

### Get the type:
- To get the type of variable ``` print(type(variable name))```

### Single or Double Quotes:
- String can be declare by single or double quotes.
- Single and double quorte are same.
### Case sensative:
- Variable names are case sensative.
``` 
z = 5 
Z = 9

Z will not overwrite z. Both variable are different
```

````python
x = str(5) # X will be '5'
y = int(5) # y will be 5
z = float(5) # z will be 5.0

print(type(x))
print(x)
print(type(y))
print(y)
print(type(z))
print(z)

x = "Hello world"
y = 'Hello world'
print(x)
print(y)
````

# Rules for variable Names:
- A variables name must be start with a letter or underscore character.
- A variables name cannot be start with number.
.- A variable name can only contain alpha-numeric character and underscore. (A4--z,0-9,_)
- A variables name is a case sensative.
- A variables name cannot be any Python Keywords

### Camel Case:
- Except the first word, each word starts with capital letter.
```
myName = "Keshab" 
```
### Pascal Case:
- Each words starts with capital letter.
```
MyName = 'Keshab'
```
### Snake Case:
- Each word is seperated by underscore (_).
```
my_variable_name = 'Keshab'
```
