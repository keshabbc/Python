# Variables
- Variable is a container used to store data values.
- In python there is no command to declare a variable.
- Variable is created when we assign a value.
- Name of variable is case sensative.
- To specify data types of a variable we use casting.
- String variable can be declare by using single or double quotes.
- To find the data type of a variable we use type() 

# Rules for variables
- A variables name must be start with a letter or underscore character.
- A variables name cannot be start with number.
.- A variable name can only contain alpha-numeric character and underscore. (A4--z,0-9,_)
- A variables name is a case sensative.
- A variables name cannot be any Python Keywords
