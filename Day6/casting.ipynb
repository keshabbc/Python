### Casting:
 - Casting means converting the data type of a variable or value into another data type.
 - casting constructor functions are used to casting.

 Casting constructor functions are:

 - int(): This function convert value  into integr.
 ```
x = int(5.9)
print(x)
```
 - float(): This function convert value  into float(decimal).
 ```
x = float(5)
print(x)
```
 - str(): This function convert value  into string.
```
x = str(100)
print(x)
```

````python
#integer

x = int(6)   # x will be 6
y = int(9.99) # y will be 9
z = int("4") # z will be 4

print(x)
print(y)
print(z)
````

````python
#float

x = float(5)     # x will be 5.0
y = float(2.7)   # y will be 2.7
z = float("8")   # z will be 8.0

print(x)
print(y)
print(z)
````

````python
#string

x = str("d5") # x will be 'd5'
y = str(43)    # y will be '43'
z = str(2.0)  # z will be '2.0'

print(x)
print(y)
print(z)
````
